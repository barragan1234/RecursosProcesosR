## Qu son los factores?
Los factores son una estructura de datos para manejar variables categricas. 

## Qu es una variable categrica?
Las variables categricas son datos que toman una cantidad finita de valores. Es decir, que pueden pertenecer a una de varias clases o categoras.

Ejemplos pueden ser rangos de edades, colores, o das de la semana

En anlisis de datos y estadstica las variables categricas se tratan de manera diferente a las variables continuas o numricas.

Necesitamos tener un objeto en R para eso. Los factores permiten trabajar con modelos y grficas de variables categricas. Por eso son importantes.

## Cmo crear factores en R
Lo primero es crear un factor a partir del vector de tallas. Para eso utilizamos la funcin factor() con el vector como argumento. Las categoras o clases de un factor se llaman niveles y podemos mirarlos utilizando la funcin levels() y pasando el factor como argumento.

En R, creamos el factor, utilizamos un nombre que nos ayude a saber el contenido de la variable.

```r
# crear factor de un vector
tallas_factor <- factor(tallas)
```

```r
tallas_factor
```

Luego hacemos la grfica y vemos como ahora podemos tener conteos de cada nivel es un avance!

```r
# graficar factor
plot(tallas_factor)
```

Hay un detalle en los niveles que quizs ya te diste cuenta Hay tallas en maysculas y minsculas.

```r
# mirar niveles de factor
levels(tallas_factor)
```

Esto ocurre porque capturamos as los datos. Pero no pasa nada, es algo bastante comn. Para corregir esto tenemos que recodificar los niveles. Veremos cmo hacerlo en la prxima seccin

# Cmo recodificar factores en R
Para cambiar el nombre de los niveles, tenemos que modificar la funcin factor(). Adems del vector, esta funcin tiene otras opciones.

Una de estas opciones es levels, que es donde escribimos los niveles que tenemos, las maysculas y minsculas la otra opcin es labels, que es donde escribimos los niveles como los queremos. En nuestro caso, decidimos tener todo en mayscula.

En R, hacemos estos cambios escribiendo los niveles y sus nuevas etiquetas, esto se llama limpieza de datos.

```r
# creando factor recodificado
tallas_recodificado <- factor(tallas,
                              levels = c("g", "m", "M", "S"),
                              labels = c("G", "M", "M", "S"))
tallas_recodificado
```

Vale la pena ver el resultado. Si hacemos la grfica otra vez, ya tiene ms sentido, muy bien!

```r
# graficando ventas_recodificado
plot(tallas_recodificado)
```

## Cmo ordenar factores en R
Sabemos que S es ms chico que M que es ms chico que G. Pero cmo hacemos para que R sepa eso?

Para decirle a R que el factor tiene algn orden tenemos que agregar la opcin orderer = TRUE. Mantenemos la recodificacin de la seccin anterior.

Volviendo a R, creamos el factor ordenado agregando el orden. despus copiamos parte de lo de arriba y cambiamos el orden en levels y labels y ya est

```r
# ordenando niveles (copiar factor anterior)
tallas_ordenado <- factor(tallas,
                          ordered = TRUE,
                          levels = c("S", "m", "M", "g"),
                          labels = c("S", "M", "M", "G"))
```

si miramos el factor vemos que ahora los niveles tienen smbolos de menor que

```r
# viendo el orden en los niveles
tallas_ordenado
```

si hacemos la grfica otra vez, vemos que ahora s se muestra en el orden que queramos Esto se ve mucho mejor! excelente!

```r
plot(tallas_ordenado)
```
